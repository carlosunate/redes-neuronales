### Fashion MNIST — Clasificación de prendas (solo capas densas)
### Características incluidas:

 - Regularización L1 + L2 en pesos (kernel_regularizer)
 - Batch Normalization
 - Dropout
 - Learning rate adaptativo (ReduceLROnPlateau)
 - EarlyStopping (con restore_best_weights=True)
 - Visualización de algunas imágenes, curvas de entrenamiento,
   evaluación en test y visualización de predicciones (con color según acierto)
   

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense, BatchNormalization, Dropout, Activation
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ---------------------------
# 0) Semillas para reproducibilidad
# ---------------------------
np.random.seed(42)
tf.random.set_seed(42)

# ---------------------------
# 1) Cargar y preprocesar Fashion MNIST
# ---------------------------
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# Nombres de las clases (para mostrar más tarde)
class_names = ['Camiseta/top', 'Pantalón', 'Suéter', 'Vestido', 'Abrigo',
               'Sandalia', 'Camisa', 'Zapatilla', 'Bolso', 'Botín']

print("Clases:", class_names)
print(f"Tamaño train: {X_train.shape}, tamaño test: {X_test.shape}")

# Normalizamos pixeles a [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Observación: vamos a usar Flatten + Dense, por eso dejamos la forma (28,28)
# (Flatten estará dentro del modelo)

# ---------------------------
# 2) Mostrar algunas imágenes de ejemplo
# ---------------------------
def mostrar_muestras(X, y, n_cols=8):
    n = n_cols
    plt.figure(figsize=(n*1.5, 2))
    indices = np.random.choice(len(X), n, replace=False)
    for i, idx in enumerate(indices):
        ax = plt.subplot(1, n, i+1)
        plt.imshow(X[idx], cmap='gray')
        plt.axis('off')
        plt.title(class_names[y[idx]], fontsize=9)
    plt.suptitle('Ejemplos de Fashion MNIST (muestras aleatorias)')
    plt.show()

mostrar_muestras(X_train, y_train, n_cols=8)

# ---------------------------
# 3) Construcción del modelo (solo Dense + BatchNorm + Dropout)
#    - Regularización L1+L2 aplicada a los kernels de las capas Dense
# ---------------------------
# ---------------------------
# Capa Dense con regularización y BatchNorm
# ---------------------------
# kernel_regularizer=reg : Aplica regularización L1+L2 a los pesos de la capa
#                          para evitar overfitting y ayudar a generalización.
# use_bias=False          : Desactiva el bias de la capa Dense porque BatchNormalization
#                          ajusta automáticamente la media de las activaciones,
#                          haciendo redundante el bias.
# La secuencia típica es: Dense -> BatchNorm -> Activation -> Dropout
# Esto permite controlar el flujo de la señal y la regularización de manera efectiva.

l1_coef = 1e-5
l2_coef = 1e-4
reg = regularizers.l1_l2(l1=l1_coef, l2=l2_coef)

model = Sequential([
    Flatten(input_shape=(28, 28)),

    # 1ª capa densa -> BatchNorm -> ReLU -> Dropout
    Dense(512, kernel_regularizer=reg, use_bias=False),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.3),

    # 2ª capa densa
    Dense(256, kernel_regularizer=reg, use_bias=False),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.3),

    # 3ª capa densa
    Dense(128, kernel_regularizer=reg, use_bias=False),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.3),

    # Capa de salida (10 clases) con softmax
    Dense(10, activation='softmax')
])

# Mostramos la arquitectura resumida
model.summary()

# ---------------------------
# 4) Compilación: optimizador Adam con lr inicial y métricas
#    Además añadimos callbacks: EarlyStopping + ReduceLROnPlateau
# ---------------------------
initial_lr = 1e-3
optimizer = tf.keras.optimizers.Adam(learning_rate=initial_lr)

# Binary/Multiclass: usamos sparse_categorical_crossentropy porque labels son enteros
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Callbacks
earlystop = EarlyStopping(monitor='val_accuracy',
                          patience=8,
                          restore_best_weights=True,
                          verbose=1)

reduce_lr = ReduceLROnPlateau(monitor='val_loss',
                              factor=0.5,
                              patience=3,
                              min_lr=1e-6,
                              verbose=1)

# ---------------------------
# 5) Entrenamiento

history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1,  # 10% del train para validación
    callbacks=[earlystop, reduce_lr],
    verbose=2
)

# ---------------------------
# 6) Curvas de entrenamiento (pérdida y accuracy)
# ---------------------------
plt.figure(figsize=(12, 4))

# Pérdida
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='loss (train)')
plt.plot(history.history['val_loss'], label='loss (val)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.title('Curva de pérdida')

# Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='accuracy (train)')
plt.plot(history.history['val_accuracy'], label='accuracy (val)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.title('Curva de accuracy')

plt.tight_layout()
plt.show()

# ---------------------------
# 7) Evaluación en el conjunto de test
# ---------------------------
loss_test, acc_test = model.evaluate(X_test, y_test, verbose=0)
print(f"\nResultado en test -> Loss: {loss_test:.4f}, Accuracy: {acc_test:.4f}")

# ---------------------------
# 8) Predicciones: mostramos varias imágenes con su etiqueta real y predicha
#    Titulos en verde si acierta, rojo si falla (texto debajo de la imagen para evitar cortes)
# ---------------------------
num_cols = 6
num_rows = 3
num = num_cols * num_rows
indices = np.random.choice(len(X_test), num, replace=False)
imgs = X_test[indices]
true_labels = y_test[indices]
preds = model.predict(imgs)
pred_labels = np.argmax(preds, axis=1)

plt.figure(figsize=(num_cols*2, num_rows*2.5))
for i in range(num):
    ax = plt.subplot(num_rows, num_cols, i+1)
    plt.imshow(imgs[i], cmap='gray')
    plt.axis('off')

    verdadero = class_names[true_labels[i]]
    predicho = class_names[pred_labels[i]]
    acierto = (true_labels[i] == pred_labels[i])

    # Título corto arriba: clase real
    ax.set_title(verdadero, fontsize=9, pad=3)

    # Texto debajo con Predicted: (en verde/rojo)
    color = 'green' if acierto else 'red'
    ax.text(0.5, -0.18, f"Pred: {predicho}", transform=ax.transAxes,
            fontsize=9, ha='center', va='top', color=color)

plt.suptitle('Predicciones sobre muestras del test (texto debajo para evitar recortes)')
plt.subplots_adjust(hspace=0.8)
plt.show()
```

## Notas finales:

 - Usamos solo capas densas (Flatten -> Dense ... -> Dense) para cumplir la restricción.
 - Regularización aplicada: L1 + L2 en kernels, BatchNormalization y Dropout.
 - LR adaptativo: ReduceLROnPlateau reduce LR cuando la val_loss no mejora.
 - EarlyStopping con restore_best_weights=True para evitar overfitting.
 - Puedes ajustar las tasas de dropout o los coeficientes L1/L2 para ver su efecto.
